# Data science visualization

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
def my_show(x,y):
    fig, ax = plt.subplots()
    ax.plot(x, y)
    plt.show()
```

```python
my_show(x=np.linspace(0, 2 * np.pi, 200),y=np.cos(x))
```

```python
my_show(x=np.linspace(0, 2 * np.pi, 200),y=np.tan(x))
```

```python
np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

```python
import matplotlib.pyplot as plt
import csv

with open('../static/rank.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # skip header row
    column2 = [[row[1],float(row[2].strip('%'))/100] for row in reader]
languages, popularity = zip(*column2)

fig, ax = plt.subplots()

ax.bar(languages, popularity)

ax.set_title('2023.5 programming language')
plt.xticks(rotation=90)
# plt.subplots_adjust(bottom=0.4)
ax.set_xlabel('languages')
ax.set_ylabel('popularity')

plt.show()
```
